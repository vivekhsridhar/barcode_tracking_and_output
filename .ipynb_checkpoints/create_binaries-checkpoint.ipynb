```python
import numpy as np
import pandas as pd
import h5py
import matplotlib.pyplot as plt
import itertools
import os
from scipy import stats
```

```python
name = '2a_2k_3x3_3ir_reconstructed'
f = h5py.File('/Users/viveksridhar/Documents/Code/Python/barcode_tracking_and_output/data/' + name + '/' + name + '.h5', "r")
```

```python
group = f['data/']
```

```python
direction = group['direction'].value
frame = group['frame'].value
ind = group['identity'].value
pos_x = group['x'].value
pos_y = group['y'].value
```

```python
ind.shape
```

```python
data = [frame, ind, pos_x, pos_y, -np.cos(direction), -np.sin(direction)]
data = np.array(data).T
```

```python
df = pd.DataFrame(data, columns=['frame', 'id', 'pos_x', 'pos_y', 'dir_x', 'dir_y'])
```

```python
f.close()
```

```python
df
```

```python
data = []

# Create individual specific dataframes
g = df.groupby("id")
d = dict(iter(g))
names = list(d.keys())
names = [int(name) for name in names]

# Removee ID column and create dataframes with IDs in their name
for i, name in enumerate(names):
    del d[name]['id']
    globals()['df_' + str(name)] = d[name]
    data.append(d[name])

# Change working directory so created binaries are placed in the data folder
os.chdir('/Users/viveksridhar/Documents/Code/Python/barcode_tracking_and_output/data/' + name + '/')

# Merge dataframes based on presence of frame in both individuals and rename columns
# Then write directions to arrays and convert them to binaries
for i in range(len(data)-1):
    for j in range(i+1,len(data)):
        # merge
        globals()['df' + str(i) + '_' + str(j)] = pd.merge(data[i], data[j], on='frame', how='inner')
        
        # rename
        globals()['df' + str(i) + '_' + str(j)].columns = ['frame', 'x' + str(i), 'y' + str(i), 'dir_x' + str(i), 'dir_y' + str(i), 'x' + str(j), 'y' + str(j), 'dir_x' + str(j), 'dir_y' + str(j)]
        
        # initialise file names
        name_x1 = 'dir' + str(i) + str(j) + '_' + 'x' + str(i)
        name_y1 = 'dir' + str(i) + str(j) + '_' + 'y' + str(i)
        name_x2 = 'dir' + str(i) + str(j) + '_' + 'x' + str(j)
        name_y2 = 'dir' + str(i) + str(j) + '_' + 'y' + str(j)
        
        # write to arrays
        globals()[name_x1] = np.array(globals()['df' + str(i) + '_' + str(j)]['dir_x' + str(i)]).astype('float32')
        globals()[name_y1] = np.array(globals()['df' + str(i) + '_' + str(j)]['dir_y' + str(i)]).astype('float32')
        globals()[name_x2] = np.array(globals()['df' + str(i) + '_' + str(j)]['dir_x' + str(j)]).astype('float32')
        globals()[name_y2] = np.array(globals()['df' + str(i) + '_' + str(j)]['dir_y' + str(j)]).astype('float32')
        
        # convert to binaries
        output_x1 = open(name_x1, 'wb')
        globals()[name_x1].tofile(output_x1)
        output_x1.close()
        
        output_y1 = open(name_y1, 'wb')
        globals()[name_y1].tofile(output_y1)
        output_y1.close()
        
        output_x2 = open(name_x2, 'wb')
        globals()[name_x2].tofile(output_x2)
        output_x2.close()
        
        output_y2 = open(name_y2, 'wb')
        globals()[name_y2].tofile(output_y2)
        output_y2.close()
```

Calculate proportion of frames where individuals are detected

```python
# Calculate total number of frames
tot = 0
for i in range(len(data)):
    tmp = max(data[i]['frame'])
    if tot < tmp:
        tot = tmp
tot

# Calculate number of frames where individual is tracked
prop = []
for i in range(len(data)):
    prop.append(len(data[i]))
prop /= tot
prop
```
