```python
import numpy as np
import pandas as pd
import h5py
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import pairwise_distances
```

```python
name = 'sunbleaks_five_free'
f = h5py.File('/Users/viveksridhar/Downloads/' + name + '.h5', "r")
```

```python
group = f['data/']
```

```python
corners = group['corners'].value
fr_idx = group['frame_idx'].value
dist = group['distances'].value
idx = group['identity'].value
```

```python
new_idx = np.where(dist < 1) #increase from 1 to get tags that aren't 100% match
```

```python
corners = corners[new_idx]
fr_idx = fr_idx[new_idx]
idx = idx[new_idx]
```

```python
ncorners = corners.reshape((-1, 2, 2, 2))
direc = ncorners.mean(axis=2)
direc = direc[:,0,:]-direc[:,1,:]
```

```python
loc = corners.mean(axis=1)
loc_x = loc[:,0]
loc_y = loc[:,1]
dir_x = direc[:,0]
dir_y = direc[:,1]
length = np.linalg.norm([dir_x, dir_y], axis=0)
dir_x = dir_x / length
dir_y = dir_y / length
#dir_deg = np.degrees(np.arctan2(dir_x,dir_y))
data = [fr_idx, idx, loc_x, loc_y, dir_x, dir_y]
data = np.array(data).T
```

```python
df = pd.DataFrame(data, columns=['frame', 'id', 'pos_x', 'pos_y', 'dir_x', 'dir_y'])
```

```python
f.close()
```

```python
for idx, ID in enumerate(np.unique(df['id'])):
    df['id'][df['id'] == ID] = idx
```

```python
df_plot = df[(df['frame'] <= 80000)&(df['frame'] > 75000)]
df_plot = df

plt.figure(figsize=(10,10))
plt.scatter(df_plot['pos_x'], df_plot['pos_y'], c=df_plot['id'].values, cmap='Set1', s=10, alpha=0.2)
plt.xlim(500,3550)
plt.ylim(0,3010)
plt.savefig(name + '.png')
plt.show()
```

### Subset df to match subseted video and create an h5 file with that data

```python
# get data from original h5 file
name = 'sunbleak_free_tracks'
f = h5py.File('/Users/viveksridhar/Documents/Work/Cluster/output/barcode_tracking/' + name + '.h5', "r")
```

```python
group = f['data/']
```

```python
corners = group['corners'].value
fr_idx = group['frame_idx'].value
dist = group['distances'].value
idx = group['identity'].value
```

```python
min_frame = 75000
max_frame = 95000
new_idx = np.where((dist < 1) & (fr_idx <= max_frame) & (fr_idx >= min_frame))
```

```python
corners = corners[new_idx]
fr_idx = fr_idx[new_idx]
dist = dist[new_idx]
idx = idx[new_idx]
```

```python
# create new h5 file
h5file = h5py.File('output.h5', 'w')

h5file.attrs.create('fps', f.attrs.get('fps'))
h5file.attrs.create('codec', f.attrs.get('codec'))
h5file.attrs.create('height', f.attrs.get('height'))
h5file.attrs.create('width', f.attrs.get('width'))
h5file.attrs.create('total_frames', max_frame - min_frame + 1)
        
data_group = h5file.create_group('data')
frame_idx_dset = data_group.create_dataset('frame_idx', shape=fr_idx.shape, dtype=np.int64, maxshape=(None,))
corners_dset = data_group.create_dataset('corners', shape=corners.shape, dtype=np.float64, maxshape=(None, 4, 2))
identity_dset = data_group.create_dataset('identity', shape=idx.shape, dtype=np.int32, maxshape=(None,))
distances_dset = data_group.create_dataset('distances', shape=dist.shape, dtype=np.int32, maxshape=(None,))
```

```python
# copy data from original file to new h5 file
frame_idx_dset[:] = fr_idx
corners_dset[:] = corners
identity_dset[:] = idx
distances_dset[:] = dist
```

```python
f.close()
h5file.close()
```
